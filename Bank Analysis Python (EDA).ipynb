```python
bins = [0,100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins = bins, labels= labels, right = False)
```

```python
df['Income Band'].value_counts().plot(kind = 'bar')
```

Examine and Distribution of unique categories in categorial columns

```python
categorials_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for cols in categorials_cols: 
    print(f"Value Counts for '{cols}':")
    display(df[cols].value_counts())
```

```python
#Bivariate Analysis

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
    plt.xticks(rotation=45)
    plt.show()
```

Histplot of value counts for different Occupation  

```python
for cols in categorials_cols:
    if cols == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[cols])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(cols)
    plt.ylabel("count")
    plt.show()
```

Numerical Analysis 

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings']
plt.figure(figsize=(15,10))
for i, cols in enumerate(numerical_cols):
    plt.subplot(2,1, i+1)
    sns.histplot(df[cols], kde=True)
    plt.title(cols)
plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']
plt.figure(figsize=(25,20))
for i, cols in enumerate(numerical_cols):
    plt.subplot(4,3, i+1)
    sns.histplot(df[cols], kde=True)
    plt.title(cols)
plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot = True, cmap = 'crest', fmt = ".2f")
plt.title("Correlation Matrix")
plt.show()
```

Insight of EDA

1. The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
